```python
import pandas as pd
import numpy as np
```

```python
mkt_customer_data = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv').drop(columns = ['Unnamed: 0'])
```

```python
mkt_customer_data
```

```python
print(f'Dataframe shape: {mkt_customer_data.shape}')
```

```python
normalized_columns = list()

for column in mkt_customer_data.columns:
    normalized_columns.append(column.lower())

mkt_customer_data.columns = normalized_columns
```

```python
mkt_customer_data.dtypes
```

```python
mkt_customer_data.isna().sum()
```

```python
from scipy import stats

state_mode = stats.mode(mkt_customer_data['state'])
mkt_customer_data['state'] = mkt_customer_data['state'].fillna(state_mode[0][0])

response_mode = stats.mode(mkt_customer_data['response'])
mkt_customer_data['response'] = mkt_customer_data['response'].fillna(response_mode[0][0])

months_since_last_claim_mean = np.mean(mkt_customer_data['months since last claim'])
mkt_customer_data['months since last claim'] = mkt_customer_data['months since last claim'].fillna(months_since_last_claim_mean)

number_open_complaints_mean = np.mean(mkt_customer_data['number of open complaints'])
mkt_customer_data['number of open complaints'] = mkt_customer_data['number of open complaints'].fillna(number_open_complaints_mean)

vehicle_class_mode = stats.mode(mkt_customer_data['vehicle class'])
mkt_customer_data['vehicle class'] = mkt_customer_data['vehicle class'].fillna(vehicle_class_mode[0][0])

vehicle_size_mode = stats.mode(mkt_customer_data['vehicle size'])
mkt_customer_data['vehicle size'] = mkt_customer_data['vehicle size'].fillna(vehicle_size_mode[0][0])

mkt_customer_data = mkt_customer_data.drop(columns = ['vehicle type'])
```
